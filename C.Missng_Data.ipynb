```python
import numpy as np
import pandas as pd 
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],   #nan means null values or incomplete data
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
df
```

```python
df.isna().sum() # show the null value as column wise with all the sum of it for large data
```

```python
df.isna().any()
```

```python
#removing misssing data it works on row wise
df.dropna()
```

```python
df.dropna(thresh=3) #give atlest theree values ex 4 didnt have more than 3 values.
```

```python
#filling the missing data
df.fillna(value  = 0)
```

```python
#filling uniquely value for each column 
values = {"B":10,"C":20,"D":30}
df.fillna(value=values)
```
